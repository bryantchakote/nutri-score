```python
from openfoodfacts import API, APIVersion, Country, Environment, Flavor

api = API(
    username=None,
    password=None,
    country=Country.world,
    flavor=Flavor.off,
    version=APIVersion.v2,
    environment=Environment.org,
)
```

```python
# Food and drinks data from the API (200*10 = 10000 retrieved products)
products = []
for i in range(10):
    results = api.product.text_search(query="", page=i+1, page_size=200)
    products += results["products"]
```

```python
# Columns of interest
columns = [
    # Product identification
    "product_name",
    
    # Nutriscore data
    "energy", "energy_points", "energy_value",
    "saturated_fat", "saturated_fat_points", "saturated_fat_value",
    "sugars", "sugars_points", "sugars_value",
    "sodium", "sodium_points", "sodium_value",
    "proteins", "proteins_points", "proteins_value",
    "fiber", "fiber_points", "fiber_value",
    "fruits_vegetables_nuts", "fruits_vegetables_nuts_points", "fruits_vegetables_nuts_value",
    "nutriscore_score", "nutriscore_grade",
    
    # Nutrient levels info (tricolor method)
    "fat_level", "salt_level", "saturated_fat_level", "sugars_level"
]
```

```python
# Nustriscore data
ns = []

for i, p in enumerate(products):
#     if i > 1053:
#         sdf = p.get("categories_hierarchy")
#         break
        
    ## Extract the product category (the first) to drop drinks
    cat = p.get("categories_hierarchy")
    if cat is not None:
        if len(cat) > 0 and ("beverage" in cat[0] or "boissons" in cat[0]):
            continue
    
    ## Additional security: drop if the quantity contains "l" (liter)
    quant = p.get("quantity")
    if quant is not None:
        if "l" in quant:
            continue
    
    # Product identification
    id_info = [p.get("product_name")]
    
    # Nutriscore data
    ns_data = p.get("nutriscore_data")
    if ns_data is None:
        ns_info = [None] * 21
    else:
        ns_info = [
            ns_data.get("energy"), ns_data.get("energy_points"), ns_data.get("energy_value"),
            ns_data.get("saturated_fat"), ns_data.get("saturated_fat_points"), ns_data.get("saturated_fat_value"),
            ns_data.get("sugars"), ns_data.get("sugars_points"), ns_data.get("sugars_value"),
            ns_data.get("sodium"), ns_data.get("sodium_points"), ns_data.get("sodium_value"),
            ns_data.get("proteins"), ns_data.get("proteins_points"), ns_data.get("proteins_value"),
            ns_data.get("fiber"), ns_data.get("fiber_points"), ns_data.get("fiber_value"),
            ns_data.get("fruits_vegetables_nuts_colza_walnut_olive_oils"), ns_data.get("fruits_vegetables_nuts_colza_walnut_olive_oils_points"), ns_data.get("fruits_vegetables_nuts_colza_walnut_olive_oils_value")
        ]
    ns_info += [p.get("nutriscore_score"), p.get("nutriscore_grade")]
                                                                                                                                                                                                                                           
    # Nutrient levels info (tricolor method)
    tri_data = p.get("nutrient_levels")
    if tri_data is None:
        tri_info = [None] * 4
    else:
        tri_info = [tri_data.get("fat"), tri_data.get("salt"), tri_data.get("saturated-fat"), tri_data.get("sugars")]
    
    # Complete product information
    product = id_info + ns_info + tri_info
    
    # Append new product to existing ones
    ns.append(product)

# Final dataset
ns = pd.DataFrame(ns, columns=columns)
```

```python
# Remove missing and duplicated values
ns.dropna(inplace=True)
ns.drop_duplicates(inplace=True)
```

```python
# Correct a format error on nutriscore
ns.loc[ns["nutriscore_score"] == "1", "nutriscore_score"] = 1
ns["nutriscore_score"] = ns["nutriscore_score"].astype(int)
```

```python
# Indice de concordance partiel
def c_p(j, H, i):
    if j < 4: # fonction à minimiser
        return int(ref.iloc[i, j] >= H[j])
    else: # fonction à maximiser
        return int(ref.iloc[i, j] <= H[j])
```

```python
i, j = 2, 0
H = [5, 2, 1, 4, 1, 3, 0]
```

```python
c_p(j, H, i)
```

```python
def C_G(H, i, k):
    C = 0
    for j in range(len(H)):
        C += k[j] * c_p(j, H, i)
    C /= sum(k)
    return C
```

```python
k = [2, 2, 2, 2, 1, 1, 1]
C_G(H, i, k)
```

```python
def H_surclasse_bi(H, i, k, s): # s pour seuil
    return C_G(H, i, k) >= s
```

```python
s = .7
H_surclasse_bi(H, i, k, s)
```

```python
nutri_scores = ["E", "D", "C", "B", "A"]
```

```python
def PessimisticMajoritySorting(H, k, s):
    for i in range(len(k)-3, 0, -1):
        if H_surclasse_bi(H, i, k, s):
            return nutri_scores[i]
    return "E"
```

```python
def OptimisticMajoritySorting(H, k, s):
    for i in range(1, 5):
        if not H_surclasse_bi(H, i, k, s):
            return nutri_scores[i-1]
    return "A"
```

```python
nutri = ns["nutriscore_grade"].str.upper()
```

```python
k = [1, 9, 4, 1, 4, 2, 1]
# k = [2]*4 + [1]*3
s = .6
opt = ns[cols].apply(lambda H: OptimisticMajoritySorting(H, k, s), axis=1)
pes = ns[cols].apply(lambda H: PessimisticMajoritySorting(H, k, s), axis=1)
```
